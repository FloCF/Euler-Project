# Problem 12

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:

$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...$$

Let us list the factors of the first seven triangle numbers:

$$1: 1\\
3: 1,3\\
6: 1,2,3,6\\
10: 1,2,5,10\\
15: 1,3,5,15\\
21: 1,3,7,21\\
28: 1,2,4,7,14,28$$

We can see that $28$ is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?


```python
import numpy as np
import timeit

start = timeit.default_timer()
max_facts = 500

tri_nr = [0]
factors = []
i = 1
nr_factors = 0


while nr_factors<=max_facts:
    tri_nr += [tri_nr[-1] + i]
    i += 1
    if (tri_nr[-1]%np.array([2,3,5,7,11,13,17])==0).all():
        factors += [np.arange(1,tri_nr[-1]+1)[tri_nr[-1]%np.arange(1,tri_nr[-1]+1)==0]]
        nr_factors = len(factors[-1])
        #print("Nummer: " + str(tri_nr[-1]) + ", # of factors: " + str(len(factors[-1])))
    
print("Solution: " + str(tri_nr[-1]) + ", # of factors: " + str(len(factors[-1])))
stop = timeit.default_timer()

print("Time: " + str(stop - start)) 
```
