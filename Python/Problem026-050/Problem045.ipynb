# Problem 45

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

$$
\text{Triangle} \;\;\;\;\;\;	T_n=n(n+1)/2 \;\;\;\;\;\;	1, 3, 6, 10, 15, ...\\
\text{Pentagonal} \;\;\;\;\;	P_n=n(3n−1)/2 \;\;\;\;	1, 5, 12, 22, 35, ...\\
\text{Hexagonal} \;\;\;\;\;\;	H_n=n(2n−1) \;\;\;\;\;\;	1, 6, 15, 28, 45, ...\\
$$

It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.


```python
final = [] 
n = 1001
while len(final)<=2:
    triangle = [int(n*(n+1)/2) for n in range(1,n)]
    pentagonal = [int(n*(3*n-1)/2) for n in range(1,n)]
    hexagonal = [int(n*(2*n-1)) for n in range(1,n)]
    final = list(set(triangle).intersection(set(pentagonal)).intersection(hexagonal))
    n += 1000

print("Solution: " + str(final[-1]))
```
